```python
# Drawing multiple shapes
import random
import ColabTurtle.Turtle as turtle
turtle.initializeTurtle()
turtle.bgcolor(255, 255, 255)
turtle.color('red')
turtle.hideturtle()
turtle.right(270)
turtle.pensize(2)

colors = ["MediumSpringGreen", "Gold", "OrangeRed", "MediumVioletRed", "MediumPurple", "LawnGreen", "DodgerBlue", "Coral"]

def draw_shape(sides):
  angle = 360 / sides
  for _ in range(sides):
    turtle.forward(75)
    turtle.right(angle)


for i in range(3, 11):
  turtle.color(random.choice(colors))
  draw_shape(i)
```

```python
# Spirograph
import random

def color():
  r = random.randint(0, 255)
  g = random.randint(0, 255)
  b = random.randint(0, 255)
  color = (r, g, b)
  return color

import ColabTurtle.Turtle as turtle

turtle.initializeTurtle()
turtle.pensize(1)
turtle.hideturtle()
turtle.bgcolor(255, 255, 255)
turtle.speed(13)

def polygon(length, n):
    for _ in range(n):
        turtle.forward(length)
        turtle.left(360 / n)

def spiro(num):
  for i in range(int(360/num)):
    turtle.color(color())
    current_heading = turtle.heading()
    turtle.setheading(current_heading + 10)
    polygon(10, 60)

spiro(5)
```

```python
import ColabTurtlePlus.Turtle as turtle
import random

y_position = [-100, -50, 0, 50, 100, 150]
turtle_color = ["red", "green", "violet", "brown", "blue", "orange"]
all_turtle = []
turtle.clearscreen()
turtle.initializeTurtle(window=(500,500))


for turtle_index in range(0, 6):
    tu = turtle.Turtle()
    tu.shape("turtle")
    tu.color(turtle_color[turtle_index])
    tu.penup()
    tu.goto(x= -230, y = y_position[turtle_index])
    all_turtle.append(tu)

user_bet = input("Which color Turtle will win the Race:\n")

if user_bet:
    game_on = True

while game_on:
    for turt in all_turtle:
        if turt.xcor() > 230:
            if user_bet == turt.pencolor():
                print(f"You Won!, The {turt.pencolor()} turtle has won the race.")
            else:
                print(f"You Lost!, The {turt.pencolor()} turtle has won the race.")
            game_on = False

        rand_dist = random.randint(20, 30)
        turt.forward(rand_dist)
```
